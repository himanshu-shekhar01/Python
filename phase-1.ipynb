## Learn Numpy
Learn Numpy from basics

```python
import numpy as np
```

### Creating Array from List 

```python
arr_1d = np.array([1,2,3,4,5])
print("1d array",arr_1d)

arr_2d = np.array([[1,2,6],[3,4,5]])
print("2d array",arr_2d)
```

### List vs Numpy array

```python
py_list = [1,2,3]
print("list multiplication : ",py_list*2)

numpy_arr = np.array([6,7,8])
print("numpy array multiplication : ",numpy_arr*2)
```

### Creating Array from scratch

```python
zeros = np.zeros((3,4))
print("Zero array: \n",zeros)

ones = np.ones((3,3))
print("ones array: \n",ones)

constant_arr = np.full((3,2),5)
print("Const array: \n",constant_arr)

random = np.random.random((2,4))
print("Random array")
print(random)

sequence = np.arange(0,10,2)
print("Sequence:\n",sequence)

```
